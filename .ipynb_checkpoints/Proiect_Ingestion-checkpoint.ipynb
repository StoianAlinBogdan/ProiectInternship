```python
#Functie helper pentru generare de fisiere .sql cu INSERT statement-uri
def SQL_INSERT_STATEMENT_FROM_DATAFRAME(SOURCE, TARGET, FILE):
    sql_texts = []
    fout = open(FILE, 'a')
    fout.truncate(0)
    for index, row in SOURCE.iterrows():       
        sql_texts.append('INSERT INTO '+TARGET+' ('+ str(', '.join(SOURCE.columns))+ ') VALUES '+ str(tuple(row.values))+';')        
    for item in sql_texts:
        fout.write(str(item)+'\n')
    fout.close()
```

```python
SQL_INSERT_STATEMENT_FROM_DATAFRAME(boroughs_df, 'BOROUGHS', 'insert_boroughs.sql')
SQL_INSERT_STATEMENT_FROM_DATAFRAME(location_info_df, 'LOCATION_INFO', 'insert_location_info.sql')
```

```python
lista_precincts = []
for index, row in df.iterrows():
    lista_precincts.append((row['ARREST_PRECINCT'], row['ARREST_BORO']))
lista_precincts = list(dict.fromkeys(lista_precincts))
lista_precincts
```

```python
lista_precincts = [list(elem) for elem in lista_precincts]
for l in lista_precincts:
    if(l[1] == 'M'):
        l[1] = 0
    elif(l[1]=='K'):
        l[1] = 1
    elif(l[1]=='B'):
        l[1] = 2
    elif(l[1]=='Q'):
        l[1] = 3
    elif(l[1]=='S'):
        l[1]=4
```

```python
l1 = [lista1[0] for lista1 in lista_precincts]
l2 = [lista1[1] for lista1 in lista_precincts]
location_df_final = pd.DataFrame()
location_df_final['LOCATION_INFO_ID'] = [x for x in range(0, len(l1))]
location_df_final['arrest_precinct'] = l1
location_df_final['BOROUGHS_boro_id'] = l2
location_df_final
```

```python
SQL_INSERT_STATEMENT_FROM_DATAFRAME(location_df_final, 'LOCATION_INFO', 'insert_location_final.sql')
```

```python

lista_initiale = []
for index, row in location_df_final.iterrows():
    if(row['BOROUGHS_boro_id'] == 0):
        lista_initiale.append('M')
    elif(row['BOROUGHS_boro_id'] == 1):
        lista_initiale.append('K')
    elif(row['BOROUGHS_boro_id'] == 2):
        lista_initiale.append('B')
    elif(row['BOROUGHS_boro_id'] == 3):
        lista_initiale.append('Q')
    elif(row['BOROUGHS_boro_id'] == 4):
        lista_initiale.append('S')
location_df_final['extra_column'] = lista_initiale
location_df_final
```
