```python
#Functie helper pentru generare de fisiere .sql cu INSERT statement-uri
def SQL_INSERT_STATEMENT_FROM_DATAFRAME(SOURCE, TARGET, FILE):
    sql_texts = []
    fout = open(FILE, 'a')
    fout.truncate(0)
    for index, row in SOURCE.iterrows():       
        sql_texts.append('INSERT INTO '+TARGET+' ('+ str(', '.join(SOURCE.columns))+ ') VALUES '+ str(tuple(row.values))+';')        
    for item in sql_texts:
        fout.write(str(item)+'\n')
    fout.close()
```

```python
SQL_INSERT_STATEMENT_FROM_DATAFRAME(boroughs_df, 'BOROUGHS', 'insert_boroughs.sql')
SQL_INSERT_STATEMENT_FROM_DATAFRAME(location_info_df, 'LOCATION_INFO', 'insert_location_info.sql')
```

**Data frame pentru PD**

```python
pd_codes_descs = df[['PD_CD', 'PD_DESC']]
#pd_codes = np.sort(pd_codes.astype(int))
pd_codes_descs['PD_CD'] = pd_codes_descs['PD_CD'].astype(int)
pd_codes_descs = pd_codes_descs.drop_duplicates()
pd_codes_descs = pd_codes_descs.sort_values(by=['PD_CD'])
pd_dict = {'pd_id': [x for x in range(0, len(pd_codes_descs))], 'pd_code': pd_codes_descs['PD_CD'], 'pd_desc': pd_codes_descs['PD_DESC']}
pd_data = pd.DataFrame(pd_dict)
SQL_INSERT_STATEMENT_FROM_DATAFRAME(pd_data, 'PD', 'insert_pd.sql')
```

**Data frame pentru KY**

```python
ky_codes_descs = df[['KY_CD', 'OFNS_DESC']]
ky_codes_descs['KY_CD'] = ky_codes_descs['KY_CD'].astype(int)
ky_codes_descs = ky_codes_descs.drop_duplicates()
ky_codes_descs = ky_codes_descs.sort_values(by=['KY_CD'])
ky_dict = {'ky_id': [x for x in range(0, len(ky_codes_descs))], 'ky_code': ky_codes_descs['KY_CD'], 'ky_desc': ky_codes_descs['OFNS_DESC']}
ky_data = pd.DataFrame(ky_dict)
SQL_INSERT_STATEMENT_FROM_DATAFRAME(ky_data, 'KY', 'insert_ky.sql')
```
